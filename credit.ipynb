```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

imorting rhe dependencies


```python
# loading the daaset to the pandas framework
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
# diatribution of legit transaction & fradulent transaction
credit_card_data['Class'].value_counts()
```

This *Dataset* is highly unbalanced
0-->Normal transcation
1-->Fraudulent transaction


```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

Under Sampling


Build a sample dataset containing the similar distributions of normal transaction and fraudulent Transactions


Number of fraudulent transaction is -->492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
credit_card_data['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the databinto features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

Split the data into Trainingdata and Testing Data

```python
X_train,X_test,Y_train,Y_test= train_test_split (X, Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic Regression model

```python
model = LogisticRegression()
```

```python
#training the loistic regression model with training data
model.fit(X_train,Y_train)
```

model evalution

accuracy score

```python
#accuracy score on training data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on training data: ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on test data: ',test_data_accuracy)
```
